```python
from befjegy import befjegy
x=befjegy.from_file("isin.conf")
def get_name(isin):
    return [element["name"] for element in x.isin if element["isin"] == isin]    

def human_format(num):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    return '%.2f%s' % (num, ['', 'k', 'mil', 'mrd', 'T', 'P'][magnitude])
```

```python
import pandas as pd
import datetime
import time

def get_name(isin):
    return [element["name"] for element in x.isin if element["isin"] == isin]    

isins=[i['isin'] for i in x.isin]
df_rates=pd.DataFrame()
df_trades=pd.DataFrame()
df_netvalue=pd.DataFrame()
for isin in isins:
    dts=[dr["date"] for dr in x.daily_rates[isin]]
    if len(dts) >0 and \
      min(dts) < time.mktime(datetime.datetime.strptime("2010.01.01.", "%Y.%m.%d.").timetuple()) and \
      max(dts) > time.mktime(datetime.datetime.strptime("2018.09.25.", "%Y.%m.%d.").timetuple()):
        df_rates[isin]=pd.Series({datetime.datetime.fromtimestamp(c['date']):c['rate'] for c in x.daily_rates[isin]})
        df_trades[isin]=pd.Series({datetime.datetime.fromtimestamp(c['date']):c['trade'] for c in x.daily_rates[isin]})
        df_netvalue[isin]=pd.Series({datetime.datetime.fromtimestamp(c['date']):c['netassetvalue'] for c in x.daily_rates[isin]})
        print isin + ": " + get_name(isin)[0]
df_rates.ffill(axis=0,inplace=True)                #fill nan with previous valid values
df_rates=df_rates.div(df_rates.iloc[-1])           #normalize rates
df_trades=df_trades.fillna(0).cumsum()             #fill nan with zero
df_netvalue.ffill(axis=0,inplace=True)
```

```python
#Largest funds...
topn=50
for ind,val in df_netvalue.iloc[-1].sort_values(ascending=False)[:topn].iteritems():
    print ind +": (" +human_format(val)+ ") "+get_name(ind)[0]
```

```python
#Fastest growing & shrinking. Period in days. Values in number of units
period=30
topn=30
print "Most growing funds..."
for ind,val in (df_trades.iloc[-1]-df_trades.iloc[-1*period]).sort_values(ascending=False)[:topn].iteritems():
    print "  " + ind + ": (" +human_format(val)+ ") "+get_name(ind)[0]
print "Most shrinking funds..."
for ind,val in (df_trades.iloc[-1]-df_trades.iloc[-1*period]).sort_values(ascending=True)[:topn].iteritems():
    print "  " + ind + ": (" +human_format(val)+ ") "+get_name(ind)[0]
    
```

```python
def fancy_dendrogram(*args, **kwargs):
    max_d = kwargs.pop('max_d', None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)

    ddata = dendrogram(*args, **kwargs)

    if not kwargs.get('no_plot', False):
        plt.title('Hierarchical Clustering Dendrogram (truncated)')
        plt.xlabel('sample index or (cluster size)')
        plt.ylabel('distance')
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y > annotate_above:
                plt.plot(x, y, 'o', c=c)
                plt.annotate("%.3g" % y, (x, y), xytext=(0, -5),
                             textcoords='offset points',
                             va='top', ha='center')
        if max_d:
            plt.axhline(y=max_d, c='k')
    return ddata
```

```python
############################################################
# Some good documents...
# https://silburt.github.io/blog/stock_correlation.html
############################################################
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage


max_d=10
corr_rates=df_rates.corr()

mask = np.zeros_like(corr_rates, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
masked_corr=corr_rates.mask(mask,0).fillna(0)

Z = linkage(masked_corr, 'average')

plt.figure(figsize=(25, 10))
labelsize=20
ticksize=15
plt.title('Hierarchical Clustering Dendrogram', fontsize=labelsize)
plt.xlabel('stock', fontsize=labelsize)
plt.ylabel('distance', fontsize=labelsize)
fancy_dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=2.,  # font size for the x axis labels
    labels = masked_corr.columns,
    truncate_mode='lastp',
    p=20,
    max_d=max_d
)
plt.yticks(fontsize=ticksize)
plt.xticks(rotation=-90, fontsize=ticksize)
plt.show()
```
